#### Grafico de lineas

```python
x = np.linspace(0, 25, 100)

y = np.sin(x)
y2 = np.cos(x)
y3 = x
y4 = x**(1/2)

plt.figure(figsize=(10,3))

plt.plot(x,y,"o",label="Funcion de seno", color="indigo", ms=3)
plt.plot(x,y2,label="Funcion del coseno", color="steelblue", lw=3)

# ! TITULO DEL GRAFICO
plt.title("Ondas", fontsize=25)

# ! TITULO DEL EJE X
plt.xlabel("Eje X", fontsize=18)

# ! TITULO DEL EJE Y
plt.ylabel("Eje Y", fontsize=18)

# ! ACTIVA LOS LABELS
plt.legend(loc="lower right", edgecolor="gray")

# ! Limite de los rangos de los graficos
plt.xlim(0,25)

plt.ylim(-1.5,1.5)

# ! Tamaño de los numeros de los ejes
plt.tick_params(axis="both", labelsize=9.5)

# todo: Añadir subgraficos
fig, axes = plt.subplots(2,2,figsize=(6,6))

# ! primer grafico
ax = axes[0][0]
ax.plot(x,y, "--", color="navy",lw=1.5)
ax.set_xlabel("$Tiempo en segundos$", fontsize=8)
ax.set_ylabel("$Seno de x$", fontsize=8)
ax.tick_params(axis="both", labelsize=8)
ax.grid(True)

# ! segundo grafico
ax = axes[1][0]
ax.plot(x,y2)

# ! tercero grafico
ax = axes[0][1]
ax.plot(x,y3)

# ! cuarto grafico
ax = axes[1][1]
ax.plot(x,y4)

# ! MOSTRAR EL GRAFICO
# ! mostrar fondo en forma de grilla
plt.grid(True)
plt.subplots_adjust(hspace=0.4, wspace=0.45)
plt.show()
```

### PARAMETROS DE CHARTS

* lw: ancho de la linea

* ms: tamaño de las esferas del grafico (en caso de tenerlas) 

* label: nombre de las lineas

* color: color de la linea 

* fontsize: tamaño de la letra

* labelsize: tamaño de las letras de los rangos

* loc: ubicacion de la leyenda del grafico

* ncol: cantidad de columnas de la leyenda

* edgecolor: color del borde de la leyenda

* colocar los labels entre $$ hace que el tipo de letra se vuelva italica

* MARCADORES: van entre comillas luego de x e y, pueden ser los siguientes:

1. --: linea punteada
2. -: linea punteada 2
3. o: puntos
4. o--: puntos con lineas
5. .: puntos
6. ,: pixeles

### PARAMETROS DE CHARTS SUBPLOTS

* hspace: espaciado en el eje y de los graficos

* wspace: espaciado en el eje x de los graficos

# Graficos de barras

```python

# ! Datos
categorias = ["A", "B", "C", "D", "E"]
valores = [5,7,3,8,4]
colores = ["steelblue", "slategrey", "teal", "silver", "slateblue"]
err = [1.2, 0.5, 0.8, 0.4, 0.15]

total = sum(valores)

porcentajes = [valor / total * 100 for valor in valores]

# ! Grafico 
plt.figure(figsize=(5,4))
plt.bar(categorias, valores, color=colores)
plt.xlabel("Categorias", fontsize=10)
plt.ylabel("Valores", fontsize=10)
plt.tick_params(axis="both", labelsize=8.5)
plt.ylim(0,9)

for i, (valor, porcentajes) in enumerate(zip(valores, porcentajes)):
    plt.text(i,valor + 0.1, f"{valor}\n{porcentajes: .1f}%",
    ha="center", fontsize=8)


plt.show()
```

* zip: se utiliza para combinar varios iterables (listas, tuplas, str) en un solo iterable de tuplas, cada tupla contendra su los elementos de las posiciones correspondientes. 

* barh: son para los graficos de barras orizontales

ej:  
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
result = zip(list1, list2)

# Si convertimos el resultado a lista:
print(list(result))  # [(1, 'a'), (2, 'b'), (3, 'c')]

Tambien pueden desempaquetarse: 

ej: x,y = zip(*result)

# x sera (1,2,3) 
# y sera (a,b,c) 


# Graficos - Histogramas

* random.normal = se utiliza para generar números aleatorios que siguen una distribución normal (también conocida como distribución gaussiana)

parametros:

1. loc = la media de la distribucion normal, 0.0 por defecto

2. scale = desviacion estandar de la distrubicion normal, controla la dispercion de los datos

3. size = tamaño de la muestra 

* random.seed = permite generar un mismo grupo de numeros aleatorios en multiples ocaciones 

* bins: controla la division de en el rango de los datos en intervalos, cuenta cuantos datos entran en cada intervalo, cuantas mas bins, mayor la cantidad de barras y mas detallados sera los graficos

* density: 
** Normaliza el histograma para que el área total bajo el histograma sea igual a 1, mostrando la densidad de probabilidad en lugar de las frecuencias absolutas. permite comparaciones entre distribuciones sin que el tamaño de los datos afecte la interpretación.

se utiliza para inicializar el generador de números aleatorios con un valor específico, conocido como "semilla". obtendrás la misma secuencia de números aleatorios cada vez que ejecutes el mismo código con la misma semilla.

* guardar graficos como imagenes = plt.savefig("nombre.formato", dpi=150)

```python

# ! semilla para reproducibilidad
np.random.seed(5)

# ! Datos
alt_mujeres = np.random.normal(157, 7, 10000)
alt_hombres = np.random.normal(170, 7, 10000)

# ! Graficos  
plt.figure(figsize=(5.5,4))
plt.hist(alt_mujeres, label="ALtura mujeres", bins=25, density=True, histtype="step")
plt.hist(alt_hombres, label="ALtura hombres", bins=25, density=True, histtype="step")

# ! Titulos y ejes
plt.title("Distribucion de alturas", fontsize=12)
plt.xlabel("Altura [cm]", fontsize=10)
plt.tick_params(axis="both", labelsize=8.5)

# ! Mostrar
plt.legend(loc="upper right", fontsize=10, edgecolor="darkgray")
plt.grid()
plt.show()
```

# Graficos de pastel

* explode = genera una separacion de esa porcion del grafico

* autopct = mostrar los valores del grafico en porcentajes

* startangle = el angulo en el cual iniciara el grafico 

```python

# ! Datos
mascotas = ["perro", "gato", "loro", "conejo", "pez"]
valores = [36,15,89,45,18]

# ! Grafico
plt.pie(valores, labels=mascotas, explode=(0,0,0.1,0,0), autopct="%1.1f%%")

# ! titulos
plt.title("Diagrama de circulos", fontsize=20)

# ! mostrar
plt.show()
```
